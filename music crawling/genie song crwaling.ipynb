```python
from selenium import webdriver
driver = webdriver.Chrome('C:/Users/sdh20/Desktop/playwithdata/chromedriver.exe')
url = 'https://www.genie.co.kr/chart/top200'
driver.get(url)
from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# 100개의 인기차트 값 가져오기
song = soup.select('tbody > tr')

import pandas as pd

song_data =[]
rank =1
for i in song:
    title = i.select('td.info > a')[0].text.replace('\n', '').replace(' ', '')
    sing = i.select('td.info > a')[1].text
    song_data.append(['genie',rank, title, sing])
    rank = rank + 1

columns = ['서비스', '순위', '타이틀', '가수']

genieDataFrame = pd.DataFrame(song_data, columns = columns)
```

```python
genieDataFrame.to_excel('genie.xlsx', index = True)
```
